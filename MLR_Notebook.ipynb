Importing all the libraries required for the data analysis and science


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
dummy = pd.get_dummies(data['State'])
```

```python
data.head(15)
```

```python
data = data.drop('State',axis=1)
```

```python
data = pd.concat([data,dummy],axis=1)
```

```python
data
```

```python
y = data.iloc[:,3]
```

```python
x = data.drop('Profit',axis=1)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x,y)
```

Let us perform the same model using different package ie statsmodel.formula.api smf


```python
import statsmodels.formula.api as smf
```

```python
data.columns = [cname.replace(' ','_') for cname in data.columns]
```

```python
data.columns = [cname.replace('&','_')for cname in data.columns]
```

```python
model = smf.ols(formula='''Profit ~ Administration + Marketing_Spend + R_D_Spend + Profit + 
                California + Florida + New_York''',data=data).fit()
```

```python
model.summary()
```
